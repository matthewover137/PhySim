```python
import math
import numpy as np
import random
import matplotlib.pyplot as plt

```

```python
numTrials = 10000
data = np.zeros(numTrials)
for i in range(numTrials):
    boxesOpened = 0
    unlocked = np.zeros((12))
    notAll = True
    boxesOpened = 0
    while(notAll):
        rand = random.randint(1,12)

        unlocked[rand - 1] += 1
        boxesOpened += 1

        if np.all(unlocked != 0):
            notAll = False
            break

    data[i] = boxesOpened

```

```python
binNum = 40
hist = np.histogram(data, bins=binNum)
plt.hist(data, bins=binNum, edgecolor='black')  

plt.title('Boxes it takes')
plt.xlabel('Num boxes it took')
plt.ylabel('Num of occurences')
plt.show()
```
